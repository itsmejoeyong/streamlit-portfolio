# Initial Setup

```python
import os

import duckdb
import streamlit as st
```

```python
con = duckdb.connect(database='duckdb/blood_donation_pipeline_v2.duckdb')
```

```python
download_dir = "load"

for files in os.listdir(download_dir):
    table_name = files.split(".")[0].replace("-", "_")
    file_path = os.path.join(download_dir, files)
    query = f"""
    CREATE OR REPLACE TABLE {table_name} AS
    SELECT * FROM '{file_path}'
    """
    con.execute(query)
    print(f"created table {table_name}")
```

```python
query = """
SELECT * FROM donations_facility LIMIT 5;
"""
con.execute(query).df()
# donations by hospital by blood_[type] 
# donations by hospital by social_[type]
# donations by hospital by donations_[frequency]
```

```python
query = """
SELECT * FROM donations_state LIMIT 5;
"""
con.execute(query).df()
# donations by state by blood_[type] 
# donations by state by social_[type]
# donations by state by donations_[frequency]
```

```python
query = """
SELECT * FROM newdonors_facility LIMIT 5;
"""
con.execute(query).df()
# donations by hospital by age group
```
